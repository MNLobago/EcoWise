```python
template = "Question:\n{question}\n\nAnswer:\n{answer}"
def colorize_text(text):
    for word, color in zip(["Question", "Answer"], ["red", "green"]):
        text = text.replace(f"\n\n{word}:", f"\n\n**<font color='{color}'>{word}:</font>**")
    return text
class GemmaQA:
    def __init__(self, max_length=512):
        self.max_length = max_length
        self.prompt = template
        self.llm = llm
        
    def query(self, question):
        response = self.llm.generate(
            self.prompt.format(
                question=question,
                answer=""), 
            max_length=self.max_length)
        display(Markdown(colorize_text(response)))
        
gemma_qa = GemmaQA()

question = "Why does too much carbon dioxide in the atmosphere cause climate change?"
gemma_qa.query(question)#1
```

```python
template = "Question:\n{question}\n\nAnswer:\n{answer}"

def colorize_text(text):
    for word, color in zip(["Question", "Answer"], ["red", "green"]):
        text = text.replace(f"\n\n{word}:", f"\n\n**<font color='{color}'>{word}:</font>**")
    return text

class GemmaQA:
    def __init__(self, max_length=512):
        self.max_length = max_length
        self.prompt = template
        self.llm = llm  # Assuming the model is already initialized.
        
    def query(self, question):
        response = self.llm.generate(
            self.prompt.format(
                question=question,
                answer=""), 
            max_length=self.max_length)
        return response

gemma_qa = GemmaQA()

# Collecting responses
results = []
for item in tqdm(NewTest1, desc="Processing Questions"):  # Adding tqdm for progress tracking
    question = item['question']
    
    # Start time
    start_time = time.time()
    
    # Get model's answer
    answer = gemma_qa.query(question)
    
    # End time
    end_time = time.time()
    
    # Calculate response time
    response_time = end_time - start_time  # Time taken in seconds
    
    results.append({
        'question': question,
        'model_answer': answer,
        'expected_answer': item['answer'],
        'category': item['category'],
        'response_time': response_time  # Add response time to results
    })

# Create a DataFrame from results
df_results = pd.DataFrame(results)

# Saving the results to a CSV file
df_results.to_csv("model_responses_with_time_tqdm2.csv", index=False)

# Display the DataFrame to verify
display(df_results)
```
